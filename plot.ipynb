#### Plot Gas and Stellar Velocity

- [SDSS Gas and Stellar Velocity tutorial](https://sdss-marvin.readthedocs.io/en/latest/tutorials/exercises/marvin_kinematics_solution.html)

- [SDSS Plotting tutorial](https://sdss-marvin.readthedocs.io/en/latest/tutorials/plotting-tutorial.html)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from marvin import config
config.setDR('DR16')

from marvin.tools.maps import Maps
import marvin.utils.plot.map as mapplot
```

```python
## plotting tutorial https://sdss-marvin.readthedocs.io/en/latest/tutorials/plotting-tutorial.html

# stellar velocity -> stellar_vel
# ionized gas component -> H_alpha 
# ionzied gas component PA -> H_alpha + NII

mangaids = ['1-113520', '1-38618', '1-38510', '1-96154']
mapnames = ['stellar_vel', 'emline_gvel_ha_6564']

with plt.style.context('seaborn-darkgrid'):
    
    rows = len(mangaids)
    cols = len(mapnames)
    fig, axes = plt.subplots(rows, cols, figsize=(10, 25))
    
    for row, mangaid in zip(axes, mangaids):
        
        maps = Maps(mangaid=mangaid, mode = 'local' )
        for ax, mapname in zip(row, mapnames):
            mapplot.plot( dapmap=maps[mapname], fig=fig, ax=ax, title=' '.join((mangaid, mapname)))
```
