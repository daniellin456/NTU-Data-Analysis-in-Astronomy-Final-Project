```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from marvin import config
config.setDR('DR16')

from marvin.tools.maps import Maps
import marvin.utils.plot.map as mapplot
plt.style.use('seaborn-darkgrid')

from marvin.utils.general.images import showImage
```

```python
## plotting tutorial https://sdss-marvin.readthedocs.io/en/latest/tutorials/plotting-tutorial.html

# stellar velocity -> stellar_vel
# ionized gas component -> H_alpha 
# ionzied gas component PA -> H_alpha + NII

plateifu = '8084-1902'
#mask, fig, ax= Maps(plateifu=plateifu, mode='local').getMap('stellar_vel')
#filename = plateifu + '_stvel.jpg'
#fig.savefig(filename)


halpha = maps.emline_gvel_ha_6564
ew_ha = maps.emline_sew_ha_6564
stvel = maps.stellar_vel
#stsig = maps.stellar_sigma
#stsig_corr = stsig.inst_sigma_correction()
with plt.style.context('seaborn-darkgrid'):

    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 11))
    halpha.plot(fig=fig, ax=axes[0, 0])
    stvel.plot(fig=fig, ax=axes[0, 1])
    ew_ha.plot(fig=fig, ax=axes[1, 0])

    #stsig_corr.plot(fig=fig, ax=axes[1, 1])
plt.savefig(plateifu+'havel_stvel_ew.png',dpi=600)
```

```python
mask, fig, axes = Maps(plateifu=plateifu, mode='local').get_bpt(snr_min=1, show_plot=False)
filename = plateifu + '_bpt.jpg'
fig.savefig(filename, dpi=600)
```

```python
image = showImage(plateifu=plateifu, show_image=False)
img = plt.imshow(image)
plt.title(str(plateifu))
plt.savefig(plateifu+'image.png',dpi=600)
```
